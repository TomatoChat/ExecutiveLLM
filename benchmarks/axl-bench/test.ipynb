```python
from src.models import Message
import os
from openai import OpenAI
from src.helpers.openai import runPrompt

runPrompt(
    client=OpenAI(api_key=os.getenv("OPENAI_API_KEY")),
    model="gpt-4o",
    maxTokens=1024,
    temperature=0.0,
    messages=[Message(
        role="user",
        content="Look on the internet what day is today"
    )],
    enableGrounding=True
)
```

```python
import axelrod as axl
import pprint
from src.strategies import CompletionLLM
from src.prompts.uncontextualized import STRATEGY_FULL_HISTORY
```

```python
testPlayer = CompletionLLM(promptTemplate=STRATEGY_FULL_HISTORY)
```

```python
players = (axl.Cooperator(), testPlayer)
match = axl.Match(players, 5)

match.play()
```

```python
players = [testPlayer, axl.Cooperator(), axl.Defector(),
           axl.TitForTat(), axl.Grudger()]
players.append(axl.Random())
tournament = axl.Tournament(players, turns=5, seed=101)
results = tournament.play()
```

**Printing summaries**

```python
summary = results.summarise()
pprint.pprint(summary)
```

```python
plot = axl.Plot(results)
p = plot.boxplot()
p.show()
```

```python
# distributions of wins for each strategy
p = plot.winplot()
p.show()
```

```python
# payoff matrix
p = plot.payoff()
p.show()
```
